```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_races = pd.read_csv("1-Formula_One/races.csv")
```

```python
#check the rows and headings 
df_races.head()
```

```python
#replace \N into NA
df_races = df_races.replace("\\N", np.nan)
```

```python
#convert date to date value using for loop
date_col = ["date", "fp1_date", "fp2_date", "fp3_date", "quali_date", "sprint_date"]

for col in date_col:
    df_races[col] = pd.to_datetime(df_races[col])
```

```python
#delete unnecessary column - the fp1, fp2, fp3, quali, and sprint data are mostly missing and therefore we decided to exclude it
df_races = df_races[["raceId", "year"]]

display(df_races)
```

```python
drivers = pd.read_csv("1-Formula_One/drivers.csv")

display(drivers)

drivers_nonnumeric = drivers.query("number.str.isnumeric() == False")
non_number_vals = pd.unique(drivers_nonnumeric["number"])

print(non_number_vals)

replace_vals = [np.nan]

drivers["number"] = drivers["number"].replace(non_number_vals, replace_vals)

drivers["number"] = pd.to_numeric(drivers["number"])

display(drivers)

df_drivers = drivers[["driverId", "forename", "surname", "nationality", "dob"]]

display(df_drivers)
```

```python
results = pd.read_csv("1-Formula_One/results.csv")

display(results)

results_nonnumeric = results.query("position.str.isnumeric() == False")
non_number_vals = pd.unique(results_nonnumeric["position"])

print(non_number_vals)

replace_vals = [np.nan]

results["position"] = results["position"].replace(non_number_vals, replace_vals)

results["position"] = pd.to_numeric(results["position"])

display(results)

df_results = results[["resultId", "raceId", "driverId", "constructorId", "grid", "positionOrder", "points"]]

display(df_results)
```

```python
df_constructors = pd.read_csv("1-Formula_One/constructors.csv")

df_constructors = df_constructors[["constructorId", "constructorRef"]]

display(df_constructors)
```

```python
#merging results.csv and drivers.csv
df_merged_drivers_results = pd.merge(left = df_results, right = df_drivers, how = "left", on = "driverId")

#merging with constructor.csv
df_merged_constructor_results = pd.merge(left = df_merged_drivers_results, right = df_constructors, how = "left", on = "constructorId")

#merging with races.csv
df_merged = pd.merge(left = df_merged_constructor_results, right = df_races, how = "left", on = "raceId")

display(df_merged)
```

```python
unique_nat = df_merged["nationality"].unique()

countries = []
total_points_lst = []
points_per_country = []
total_race = []


for country in unique_nat: 
    df_country = df_merged.query("nationality == @country")

    total_races = len(df_country)
    total_points = df_country["points"].sum()
    points_per_race = total_points/total_races

    countries.append(country)
    total_race.append(total_races)
    total_points_lst.append(total_points)
    points_per_country.append(points_per_race)

country_points = pd.DataFrame({
    "country": countries,
    "total_race": total_race,
    "total_points": total_points_lst,
    "points_per_race": points_per_country
})

display(country_points)

```

```python
unique_dri = df_merged["driverId"].unique()

drivers = []
total_points_lst = []
average_point_driver = []
total_race = []

for driver in unique_dri: 
    df_per_driver = df_merged.query("driverId == @driver")

    total_races = len(df_per_driver)
    total_points = df_per_driver["points"].sum()
    points_per_driver = total_points / total_races

    drivers.append(driver)
    total_race.append(total_races)
    total_points_lst.append(total_points)
    average_point_driver.append(points_per_driver)

driver_points = pd.DataFrame({
    "driver": drivers,
    "total_race": total_race,
    "total_points": total_points_lst,
    "average_points": average_point_driver
})

driver_points = driver_points.sort_values("average_points", ascending=False)

display(driver_points)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(driver_points["total_race"], driver_points["average_points"])

plt.xlabel("Total Races")
plt.ylabel("Average Points")
plt.title("Relationship Between Total Races and Average Points per Driver")

plt.show()
```
